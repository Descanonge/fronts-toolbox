# Belkin-O'Reilly Algorithm

## Ideal Jet

A idealized meandering jets with rings, as a 2D numpy array (embeded in Xarray) with swap noise and spikes.

```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

from fronts_toolbox._fields import ideal_jet, sample, add_spikes, swap_noise
from fronts_toolbox.filters.boa import boa_xarray
```

```python
sst_raw = ideal_jet()
vmin = np.min(sst_raw)
vmax = np.max(sst_raw)
sst = xr.DataArray(add_spikes(swap_noise(sst_raw)), name="sst", dims=("lat", "lon"))
filtered = boa_xarray(sst)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 2.3), layout="constrained", sharex=True, sharey=True, dpi=200)

im_kw = dict(cmap="inferno", add_labels=False, center=False, vmin=vmin, vmax=vmax)
sst.plot.imshow(ax=axes[0], **im_kw)
filtered.plot.imshow(ax=axes[1], **im_kw)

axes[0].set_title("SST", weight="bold")
axes[1].set_title("Filtered", weight="bold")

for ax in axes:
    ax.set_aspect("equal")
```

## MODIS Data

SST off the Californian coast, as a Dask array embeded in Xarray.

```python
sst = (
    sample("MODIS")
    .sst4
    .isel(time=0)
    .sel(lat=slice(20, 10), lon=slice(-110, -100))
    .chunk(lat=256, lon=256)
)
filtered = boa_xarray(sst)

vmin = float(sst.min())
vmax = float(sst.max())

fig, axes = plt.subplots(1, 2, figsize=(6, 2.2), layout="constrained", sharex=True, sharey=True, dpi=200)

im_kw = dict(
    add_labels=False,
    center=False,
    cbar_kwargs=dict(location="right", pad=0.05),
    vmin=vmin,
    vmax=vmax,
)
sst.plot.imshow(ax=axes[0], **im_kw)
filtered.plot.imshow(ax=axes[1], **im_kw)
axes[0].set_title("SST", weight="bold")
axes[1].set_title("Filtered", weight="bold")

for ax in axes:
    ax.set_aspect("equal")
```
